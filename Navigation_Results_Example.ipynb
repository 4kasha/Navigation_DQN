## Summary

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['font.size'] = 10

%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
def plot_scores(scores, filename, rolling_window=20):
    """Plot scores and optional rolling mean using specified window."""
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(np.arange(len(scores)), scores, color='skyblue')
    ax.axhline(y=13,linestyle='--', color='gray')
    rolling_mean = pd.Series(scores).rolling(rolling_window).mean()
    plt.plot(rolling_mean, color='m')
    plt.ylabel('Score')
    plt.xlabel('Episode #')
    if filename:
        plt.savefig(filename, transparent=True, dpi=300)

def plot_scores_multi(scores, filename, rolling_window=20):
    rolling_mean=[[] for _ in range(len(scores))]
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.axhline(y=13,linestyle='--', color='gray')
    for i in range(len(scores)):
        rolling_mean[i] = pd.Series(scores[i]).rolling(rolling_window).mean()
    plt.plot(rolling_mean[0], label="DQN")
    plt.plot(rolling_mean[1], label="DDQN")
    plt.plot(rolling_mean[2], label="Dueling + DQN")
    plt.ylabel('Mean score')
    plt.xlabel('Episode #')
    plt.legend(borderaxespad=1)
    plt.savefig(filename, transparent=True, dpi=300)
    
```

```python
scores=[]

f = open("scores/scores_DQN.txt","rb")
scores = [pickle.load(f)]
f = open("scores/scores_DDQN.txt","rb")
scores.append(pickle.load(f))
f = open("scores/scores_Dueling_DQN.txt","rb")
scores.append(pickle.load(f))
```

```python
plot_scores_multi(scores, "summary.png", rolling_window=100)
```

```python
plot_scores(scores[0], "DQN.png", rolling_window=100)
```

```python
plot_scores(scores[1], "DDQN.png", rolling_window=100)
```

```python
plot_scores(scores[2], "Dueling_DQN.png", rolling_window=100)
```
